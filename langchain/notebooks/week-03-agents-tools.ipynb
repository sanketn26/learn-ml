Agent workflow:
1. Think: "I need two tools: get_account_balance and get_billing_date"
2. Act: Call both tools

4. Think: Format response
5. Return: "Your balance is $X. Next billing: Y"

## üìö Key Concepts

- **Agents**: LLMs that make tool decisions
- **Tools**: Functions agents can call (APIs, databases, calculators)
- **ReAct**: Alternates between reasoning and acting
- **Toolkits**: Pre-built tool collections (Google, Wikipedia, SQL)
- **Error handling**: Gracefully recover from tool failures
